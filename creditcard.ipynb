```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
#0--->Normal Transaction
#1---->Fraudulent Transaction
print(credit_card_data['Class'].value_counts())
```

```python
#seperating the data for analysis

legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]

```

```python
#under Sampling
#Built a sample dataset containing similar distribution of model
legit_sample=legit.sample(n=492)
```

```python
#concatenate 2 data frames
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
import matplotlib.pyplot as plt
class_counts = new_data['Class'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(class_counts, labels=['Non-Fraud (0)', 'Fraud (1)'], autopct='%1.1f%%', colors=['skyblue', 'red'])
plt.title('Fraud vs. Non-Fraud Transactions')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.scatterplot(data=new_data, x='Time', y='Amount', hue='Class', palette={0: 'blue', 1: 'red'}, alpha=0.5)

# Labels and title
plt.xlabel("Time (seconds)")
plt.ylabel("Transaction Amount")
plt.title("Scatter Plot of Transaction Amount vs. Time")
plt.legend(title="Class", labels=["Non-Fraud (0)", "Fraud (1)"])
plt.show()
```

```python
new_data.groupby('Class').mean()
```

```python
#splitting the data into features and target
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
Y.head
```

```python
#train test split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
#stratify to evenly distribute the data
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
#model training
from sklearn.ensemble import RandomForestClassifier

# Initialize and train the model
model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
model.fit(X_train, Y_train)


```

```python
#Evaluation
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on training data :',training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data :',test_data_accuracy)
```
